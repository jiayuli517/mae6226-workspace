# 2. Shapes created by the Joukowski transformation

```python
#Start by writing a Python function
def get_function(x,y):
    z=x+y*1j
    xi=z+c**2/z
    return xi
```

```python
c=1.0
N=100
theta = numpy.linspace(0, 2*numpy.pi, N)

#On the z-plane, place a circle with radius R1=1.5, c=1, centered at the origin
R1=1.5
a1,b1=0,0
```

```python
def get_x_y(R,a,b,theta):
    x=a+R*numpy.cos(theta)
    y=b+R*numpy.sin(theta)
    return x,y
```

```python
x1,y1=get_x_y(R1,a1,b1,theta)
```

```python
numpy.shape(x1)
```

```python
xi_1 = get_function (x1,y1)
```

# 4. Flow over a symmetric Joukowski airfoil with zero angle of attack

```python
u_inf = 1.0
R4=1.15
kappa = R4**2*2*numpy.pi*u_inf
```

```python
kappa
```

```python
def get_velocity_doublet(strength, xd, yd, x, y):
    
    u = - strength/(2*math.pi)*((x-xd)**2-(y-yd)**2)/((x-xd)**2+(y-yd)**2)**2
    v = - strength/(2*math.pi)*2*(x-xd)*(y-yd)/((x-xd)**2+(y-yd)**2)**2
    
    return u, v

def get_stream_function_doublet(strength, xd, yd, x, y):
    
    psi = - strength/(2*math.pi)*(y-yd)/((x-xd)**2+(y-yd)**2)
    
    return psi
```

```python
u_doublet, v_doublet = get_velocity_doublet(kappa, a4, b4, x4, y4)
psi_doublet = get_stream_function_doublet(kappa, a4, b4, x4, y4)
```

```python
u_freestream = u_inf * numpy.ones((N_theta, N_r), dtype=float)
v_freestream = numpy.zeros((N_theta, N_r), dtype=float)

psi_freestream = u_inf * y4
```

```python
numpy.shape(u_freestream)
```

```python
# superposition of the doublet on the freestream flow
u = u_freestream + u_doublet
v = v_freestream + v_doublet
psi = psi_freestream + psi_doublet

x5,y5=get_x_y(R4,a4,b4,theta)

size = 5
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.xlabel('z-plane',fontsize=16)
pyplot.contour(x4, y4,psi, levels=numpy.linspace(x_start, x_end, 50), colors='#CD2305', linewidths=1, linestyles='solid')
pyplot.scatter(x5,y5,color='#CD2305',s=1);
```

```python
xi_5=get_function(x4,y4)
xi_6= get_function(x5,y5)
```

```python
size = 5
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.xlabel('xi-plane',fontsize=16)
pyplot.contour(xi_5.real, xi_5.imag, psi, levels=numpy.linspace(x_start, x_end, 50), colors='#CD2305', linewidths=2, linestyles='solid')
```

```python
z = x4 + y4*1j
W = (u - v*1j)/(1-(c/z)**2)
```

```python
size=10
x_start,x_end=-3,3
y_start,y_end=-3,3
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.quiver(x4,y4,u,v,units='xy',pivot='tip',width=0.005,scale=40)
```

# 6. Flow over a symmetric Joukowski airfoil with non-zero angle of attack, with circulation

```python
gamma = -1*4*math.pi*1.15*numpy.sin(-20*math.pi/180)
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
 
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u, v

def get_stream_function_vortex(strength, xv, yv, X, Y):

    psi = strength/(4*math.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    
    return psi
```

```python
u_vortex, v_vortex = get_velocity_vortex(gamma, 0, 0, x4, y4)
psi_vortex = get_stream_function_vortex(gamma, 0, 0, x4, y4)

u = u + u_vortex
v = v + v_vortex
psi = psi + psi_vortex
```
